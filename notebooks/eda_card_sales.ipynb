```python
def get_unique_categories():
    all_categories = set()
    for df in [sales_2023_sexage, sales_2024_sexage, sales_2023_time, sales_2024_time, sales_2023_month, sales_2024_month]:
        all_categories.update(df['분류'].unique())
    return sorted(all_categories)

unique_categories = get_unique_categories()
print(f"총 업종 분류 개수: {len(unique_categories)}")
print("예시 업종 분류:", unique_categories[:])
```

```python
# 연령대별 업종 매출 비율 확인 (청년 = 20대, 30대)
def check_youth_preferred_categories(df):
    if '매출금액합계' in df.columns:
        amount_col = '매출금액합계'
    elif '매출합계(원)' in df.columns:
        amount_col = '매출합계(원)'
    else:
        raise ValueError("매출 합계 관련 컬럼을 찾을 수 없습니다.")

    youth = df[df['연령대'].isin(['20대', '30대'])]
    total = youth.groupby('분류')[amount_col].sum().sort_values(ascending=False)
    share = total / total.sum()
    return share

youth_cat_2023 = check_youth_preferred_categories(sales_2023_sexage)
youth_cat_2024 = check_youth_preferred_categories(sales_2024_sexage)

print("🔍 2023 청년 업종 소비 비중:")
print(youth_cat_2023)

print("\n🔍 2024 청년 업종 소비 비중:")
print(youth_cat_2024)
```

```python
# 야간 소비 시간대 정의
def check_night_categories(df):
    if '매출금액합계' in df.columns:
        amount_col = '매출금액합계'
    elif '매출합계(원)' in df.columns:
        amount_col = '매출합계(원)'
    else:
        raise ValueError("매출 합계 관련 컬럼을 찾을 수 없습니다.")

    night = df[(df['시간대'] >= 18) | (df['시간대'] < 6)]
    total = night.groupby('분류')[amount_col].sum().sort_values(ascending=False)
    share = total / total.sum()
    return share


night_cat_2023 = check_night_categories(sales_2023_time)
night_cat_2024 = check_night_categories(sales_2024_time)

print("🌙 2023 야간 소비 업종 비중:")
print(night_cat_2023)

print("\n🌙 2024 야간 소비 업종 비중:")
print(night_cat_2024)
```
